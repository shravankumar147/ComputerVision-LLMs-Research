```python
from smolagents import CodeAgent, DuckDuckGoSearchTool, HfApiModel

agent = CodeAgent(tools=[DuckDuckGoSearchTool()], model=HfApiModel())

agent.run("How many seconds would it take for a leopard at full speed to run through Pont des Arts?")
```

```python
query = "Plan a trip to Goa, from Hyderabad, India, Feb 1 to Feb 8. Allocate time according to number of public attraction in each, and optimize for distance and travel time. Bring all the public transportation options."
```

```python
agent.run(query)
```
