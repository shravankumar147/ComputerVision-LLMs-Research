```python
url = "https://books.toscrape.com/"
```

```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
content = response.content
```

```python
soup = bs(content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
all_articles = soup.find_all('h3')
```

```python
for article in all_articles:
    print(article.text)
```

```python
for article in all_articles:
    for a in article:
        print(a.get_attribute_list('title')[0])
```
