# MarkItDown
MarkItDown is a utility for converting various files to Markdown (e.g., for indexing, text analysis, etc). It supports:
```
PDF
PowerPoint
Word
Excel
Images (EXIF metadata and OCR)
Audio (EXIF metadata and speech transcription)
HTML
Text-based formats (CSV, JSON, XML)
ZIP files (iterates over contents)
```
To install MarkItDown, use pip: pip install markitdown. Alternatively, you can install it from the source: pip install -e .

````python
import wget

# URL of the paper
url = "https://arxiv.org/pdf/1706.03762"

# Download the file
filename = wget.download(url, out="attention_is_all_you_need.pdf")
print(f"File saved as {filename}")
````

````python
from markitdown import MarkItDown

md = MarkItDown()
result = md.convert("attention_is_all_you_need.pdf")
print(result.text_content)
````

````python
from IPython.display import Markdown, display
````

````python
display(Markdown(result.text_content))
````
